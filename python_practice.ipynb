```python
"""
Python Coding Practice — Challenge 1

Task:
Write a function called `count_vowels(text)` that returns a dictionary counting
how many times each vowel appears in the input string.

Requirements:
1. Count vowels only: a, e, i, o, u
2. Treat uppercase and lowercase the same
3. Ignore all non-vowel characters
4. Return only vowels that appear at least once
5. The dictionary keys should be lowercase vowels

Examples:
count_vowels("Hello World") -> {'e': 1, 'o': 2}
count_vowels("AEIOU!!!") -> {'a': 1, 'e': 1, 'i': 1, 'o': 1, 'u': 1}
count_vowels("rhythm") -> {}

Write your code below.
"""


def count_vowels(text):

    text = text.lower()
    
    vowel_ct_dict = {}

    vowels = ['a','e','i','o','u']

    for vowel in vowels:
        counter = 0
        if vowel in text:
            for char in text:
                if char == vowel:
                    counter += 1
            vowel_ct_dict[vowel] = counter
        else:
            continue
    return vowel_ct_dict


# Optional: use these to test your function
print(count_vowels("Hello World"))
print(count_vowels("AEIOU!!!"))
print(count_vowels("rhythm"))
print(count_vowels("aaAA"))
```

```python
"""
Python Coding Practice — Challenge 4

Task:
Write a function called `merge_counts(dict1, dict2)` that merges two dictionaries
whose values are integer counts.

Requirements:
1. Return a new dictionary
2. Do not modify dict1 or dict2
3. If a key exists in both dictionaries, add the counts together
4. If a key exists in only one dictionary, keep that count
5. The function should work with empty dictionaries

Examples:
merge_counts({'a': 2, 'b': 3}, {'a': 4, 'c': 1}) -> {'a': 6, 'b': 3, 'c': 1}
merge_counts({}, {'x': 5}) -> {'x': 5}
merge_counts({'x': 2}, {}) -> {'x': 2}
merge_counts({}, {}) -> {}

Write your code below.
"""


def merge_counts(dict1, dict2):

    merged_dict = {}

    for key, value in dict1.items():
        if key in dict2:
            merged_dict[key] = value + dict2[key]
        else:
            merged_dict[key] = value

    for key, value in dict2.items():
        if key not in dict1:
            merged_dict[key] = value

    return merged_dict


# Optional: use these to test your function
print(merge_counts({'a': 2, 'b': 3}, {'a': 4, 'c': 1}))
print(merge_counts({}, {'x': 5}))
print(merge_counts({'x': 2}, {}))
print(merge_counts({}, {}))
```

```python
"""
Python Coding Practice — Challenge 9

Task:
Write a function called `invert_dict(d)` that swaps the keys and values of a dictionary.

Requirements:
1. Return a new dictionary
2. Do not modify the input dictionary
3. Assume all values in the input dictionary are unique
4. The original values become keys in the new dictionary
5. The original keys become values in the new dictionary

Examples:
invert_dict({'a': 1, 'b': 2}) -> {1: 'a', 2: 'b'}
invert_dict({'x': 'apple', 'y': 'banana'}) -> {'apple': 'x', 'banana': 'y'}
invert_dict({}) -> {}

Write your code below.
"""


def invert_dict(d):
    inverted_dict = {}

    for key, value in d.items():

        if value not in inverted_dict.keys():
            inverted_dict[value] = [key]
        else:
            inverted_dict[value].append(key)

    return inverted_dict


# Optional: use these to test your function
print(invert_dict({'a': 1, 'b': 2}))
print(invert_dict({'x': 'apple', 'y': 'banana'}))
print(invert_dict({}))
```

```python
"""
Python Coding Practice — Challenge 10

Task:
Write a function called `word_frequencies(sentence)` that returns a dictionary
counting how many times each word appears in a sentence.

Requirements:
1. Treat uppercase and lowercase as the same
2. Ignore punctuation
3. Split words on whitespace
4. Return dictionary keys as lowercase words
5. If the input string is empty or contains only punctuation/spaces, return {}

Examples:
word_frequencies("Hello hello world!") -> {'hello': 2, 'world': 1}
word_frequencies("Cat, dog. cat") -> {'cat': 2, 'dog': 1}
word_frequencies("!!!") -> {}
word_frequencies("") -> {}

Write your code below.
"""


def word_frequencies(sentence):

    word_ct_dict = {}

    sentence = sentence.lower()

    sentence = re.sub(r'[^A-Za-z0-9|\s+]', "", sentence)

    if len(sentence.strip()) == 0:
        return {}
    
    sentence_tokens = sentence.split()

    seen = []
    
    for token in sentence_tokens:
        if token not in seen:
            pattern = r'\b' + re.escape(token) + r'\b'
            token_ct = len(re.findall(pattern, sentence))
            word_ct_dict[token] = token_ct
            seen.append(token)

    return word_ct_dict


# or: word_ct_dict[token] = word_ct_dict.get(token, 0) + 1

# Optional: use these to test your function
print(word_frequencies("Hello hello world!"))
print(word_frequencies("Cat, dog. cat"))
print(word_frequencies("!!!"))
print(word_frequencies(""))
```

```python
"""
Python Coding Practice — Challenge 11

Task:
Write a function called `most_common_word(sentence)` that returns the most
frequent word in a sentence.

Requirements:
1. Treat uppercase and lowercase as the same
2. Ignore punctuation
3. Split words on whitespace
4. Return the word as a lowercase string
5. If there is a tie, return the word that appears first in the cleaned sentence
6. If the input string is empty or contains only punctuation/spaces, return None

Examples:
most_common_word("Hello hello world!") -> 'hello'
most_common_word("Cat, dog. cat") -> 'cat'
most_common_word("dog cat dog cat") -> 'dog'
most_common_word("!!!") -> None
most_common_word("") -> None

Write your code below.
"""


def most_common_word(sentence):
    
    sentence = sentence.lower()

    sentence = re.sub(r'[^A-Za-z0-9|\s+]', "", sentence)

    if len(sentence.strip()) == 0:
        return None

    return Counter([token for token in sentence.split()]).most_common(1)[0][0]


# Optional: use these to test your function
print(most_common_word("Hello hello world!"))
print(most_common_word("Cat, dog. cat"))
print(most_common_word("dog cat dog cat"))
print(most_common_word("!!!"))
print(most_common_word(""))
```

```python
"""
Python Coding Practice — Challenge 12

Task:
Write a function called `valid_parentheses(s)` that returns True if every opening
parenthesis has a matching closing parenthesis in the correct order.

Requirements:
1. Only evaluate normal parentheses: ( and )
2. Ignore all other characters
3. Return True if parentheses are balanced
4. Return False if a closing parenthesis appears before a matching opening one
5. An empty string should return True

Examples:
valid_parentheses("()") -> True
valid_parentheses("(())") -> True
valid_parentheses("(()") -> False
valid_parentheses(")(") -> False
valid_parentheses("a(b)c") -> True
valid_parentheses("a)b(c") -> False
valid_parentheses("") -> True

Write your code below.
"""


def valid_parentheses(s):
    
    s = re.sub(r'[^()]', "", s)

    if s.startswith(")") or s.endswith("("):
        return False
    
    if len(s.strip()) == 0:
        return True

    open_pattern = re.escape("(")

    open_parenth_ct = len(re.findall(open_pattern, s))

    closed_pattern = re.escape(")")

    closed_parenth_ct = len(re.findall(closed_pattern, s))

    if open_parenth_ct != closed_parenth_ct:
        return False

    open_idx_list = []
    closed_idx_list = []

    for i, char in enumerate(s):

        if char == "(":
            open_idx_list.append(i)
        if char == ")":
            closed_idx_list.append(i)

    open_idx_sum = sum(open_idx_list)
    closed_idx_sum = sum(closed_idx_list)

    if open_idx_sum >= closed_idx_sum:
        return False

    while open_idx_list and closed_idx_list:
        if min(open_idx_list) > min(closed_idx_list):
            return False

        open_idx_list = open_idx_list[1:]
        closed_idx_list = closed_idx_list[1:]

    return True


# Optional: use these to test your function
print(valid_parentheses("()"))
print(valid_parentheses("(())"))
print(valid_parentheses("(()"))
print(valid_parentheses(")("))
print(valid_parentheses("a(b)c"))
print(valid_parentheses("a)b(c"))
print(valid_parentheses(""))
```

```python
"""
Python Coding Practice — Challenge 13

Task:
Write a function called `valid_brackets(s)` that returns True if brackets are
balanced and correctly nested.

Bracket types to evaluate:
    ()
    []
    {}

Requirements:
1. Evaluate three bracket types: parentheses (), square brackets [], and curly braces {}
2. Ignore all other characters
3. Return True only when every opening bracket has the correct matching closing bracket
4. Return False if brackets close in the wrong order
5. Return False if there are unmatched opening brackets left over
6. An empty string should return True

Examples:
valid_brackets("()") -> True
valid_brackets("([])") -> True
valid_brackets("([)]") -> False
valid_brackets("{[()]}") -> True
valid_brackets("{[(])}") -> False
valid_brackets("abc(def)[ghi]") -> True
valid_brackets("") -> True

Hint:
Use a stack.

General stack idea:
- When you see an opening bracket, push it onto the stack.
- When you see a closing bracket, check whether it matches the most recent opening bracket.
- If it matches, remove the opening bracket from the stack.
- If it does not match, return False.
- At the end, return True only if the stack is empty.

Write your code below.
"""


def valid_brackets(s):
    
    if s == "":
        return True

    open_brackets = []

    for char in s:
        if char in ['(','[','{']:
            open_brackets.append(char)

        if char in [')',']','}']:
            if not open_brackets:
                return False
            elif char == ')' and open_brackets[-1] != '(':
                return False
            elif char == ']' and open_brackets[-1] != '[':
                return False
            elif char == '}' and open_brackets[-1] != '{':
                return False
            else:
                open_brackets = open_brackets[:-1]

    if len(open_brackets) != 0:
        return False

    return True

    # if re.sub(r'[^()[]{}]', "", s)[0] in [')',']','}']:
    #     return False


# Optional: use these to test your function
print(valid_brackets("()"))              # True
print(valid_brackets("([])"))            # True
print(valid_brackets("([)]"))            # False
print(valid_brackets("{[()]}"))          # True
print(valid_brackets("{[(])}"))          # False
print(valid_brackets("abc(def)[ghi]"))   # True
print(valid_brackets(""))                # True
```

```python
"""
Python Coding Practice — Challenge 14

Task:
Write a function called `compress_string(s)` that compresses consecutive repeated
characters by writing the character followed by its count.

Requirements:
1. Count consecutive runs of the same character
2. Return the compressed string
3. If a character appears once consecutively, still include the count 1
4. Preserve uppercase/lowercase differences
5. If the input string is empty, return an empty string

Examples:
compress_string("aaabbc") -> "a3b2c1"
compress_string("abcd") -> "a1b1c1d1"
compress_string("AAaa") -> "A2a2"
compress_string("") -> ""
compress_string("!!!!!") -> "!5"

Write your code below.
"""


def compress_string(s):
    
    if s == "":
        return ""

    counter = 0
    prev_char = None
    compressed_string = ""

    for i, char in enumerate(s):
        if prev_char is None:
            prev_char = char
            counter += 1
        elif char == prev_char and (len(s) - 1) - i != 0:
            counter += 1
        elif char == prev_char and (len(s) - 1) - i == 0:
            counter += 1
            compressed_string = compressed_string + prev_char + str(counter)
        elif char != prev_char and (len(s) - 1) - i != 0:
            compressed_string = compressed_string + prev_char + str(counter)
            counter = 1
            prev_char = char
        else:
            compressed_string = compressed_string + prev_char + str(counter) + char + str(1)
        
    return compressed_string

# Optional: use these to test your function
print(compress_string("aaabbc"))
print(compress_string("abcd"))
print(compress_string("AAaa"))
print(compress_string(""))
print(compress_string("!!!!!"))
```

```python
"""
Python Coding Practice — Challenge 15

Task:
Write a function called `is_anagram(s1, s2)` that returns True if two strings are
anagrams of each other and False otherwise.

Requirements:
1. Treat uppercase and lowercase as the same
2. Ignore spaces
3. Ignore punctuation and other non-alphanumeric characters
4. Return True if both cleaned strings contain the same characters with the same counts
5. Return False otherwise

Examples:
is_anagram("listen", "silent") -> True
is_anagram("Triangle", "Integral") -> True
is_anagram("Dormitory", "Dirty room!!") -> True
is_anagram("hello", "world") -> False
is_anagram("aabb", "ab") -> False
is_anagram("", "!!!") -> True

Write your code below.
"""


def is_anagram(s1, s2):
    
    s1 = re.sub(r'[^A-Za-z0-9]', "", s1).lower()
    s2 = re.sub(r'[^A-Za-z0-9]', "", s2).lower()

    s1_dict = {}
    s2_dict = {}

    for char in s1:
        if char in s1_dict.keys():
            continue

        pattern = re.escape(char)
        matches = re.findall(pattern, s1)
        char_ct = len(matches)
        s1_dict[char] = char_ct

    for char in s2:
        if char in s2_dict.keys():
            continue

        pattern = re.escape(char)
        matches = re.findall(pattern, s2)
        char_ct = len(matches)
        s2_dict[char] = char_ct

    if s1_dict == s2_dict:
        return True

    return False

# Optional: use these to test your function
print(is_anagram("listen", "silent"))
print(is_anagram("Triangle", "Integral"))
print(is_anagram("Dormitory", "Dirty room!!"))
print(is_anagram("hello", "world"))
print(is_anagram("aabb", "ab"))
print(is_anagram("", "!!!"))
```

```python
"""
Python Coding Practice — Challenge 16

Task:
Write a function called `two_sum(nums, target)` that returns the indices of two
numbers in the list that add up to the target.

Requirements:
1. Return a list containing exactly two indices
2. The two indices must be different
3. Return the first valid pair found while scanning left to right
4. If no valid pair exists, return None
5. The input list may contain duplicate numbers

Examples:
two_sum([2, 7, 11, 15], 9) -> [0, 1]
two_sum([3, 2, 4], 6) -> [1, 2]
two_sum([3, 3], 6) -> [0, 1]
two_sum([1, 2, 3], 10) -> None

two_sum([1, 5, 1, 5], 6) -> [0, 1]
Because when scanning left to right, index 0 and index 1 form the first valid pair.

Write your code below.
"""


def two_sum(nums, target):
    
    nums_clean = [x for x in nums if int(x) or float(x) or x == 0]

    i_skip = 0

    while i_skip <= len(nums_clean) - 1:
        for i, num in enumerate(nums_clean):
            if i == i_skip:
                continue

            if nums_clean[i_skip] + nums_clean[i] == target:
                return [i_skip, i]
        i_skip += 1

    return None


# Optional: use these to test your function
print(two_sum([2, 7, 11, 15], 9))      # [0, 1]
print(two_sum([3, 2, 4], 6))           # [1, 2]
print(two_sum([3, 3], 6))              # [0, 1]
print(two_sum([1, 2, 3], 10))          # None
print(two_sum([1, 5, 1, 5], 6))        # [0, 1]

# this version conducts duplicate checks - less efficient.
```

```python
"""
Python Coding Practice — Challenge 26

Task:
Write a function called `customer_summary(transactions)` that takes a list of
transaction dictionaries and returns a dictionary summarizing each customer.

Requirements:
1. `transactions` is a list of dictionaries
2. Each dictionary has keys: "customer" and "amount"
3. Return a dictionary where each key is a customer name
4. Each value should be another dictionary with:
   - "count": number of transactions for that customer
   - "total": total amount spent by that customer
   - "average": average transaction amount for that customer
5. Preserve dictionary insertion order based on the first time each customer appears
6. Do not round the average
7. If the input list is empty, return an empty dictionary

Examples:
customer_summary([
    {"customer": "Ana", "amount": 25},
    {"customer": "Ben", "amount": 50},
    {"customer": "Ana", "amount": 15}
]) -> {
    "Ana": {"count": 2, "total": 40, "average": 20.0},
    "Ben": {"count": 1, "total": 50, "average": 50.0}
}

customer_summary([
    {"customer": "Mia", "amount": 100},
    {"customer": "Leo", "amount": 50},
    {"customer": "Mia", "amount": 25}
]) -> {
    "Mia": {"count": 2, "total": 125, "average": 62.5},
    "Leo": {"count": 1, "total": 50, "average": 50.0}
}

customer_summary([]) -> {}

Write your code below.
"""


def customer_summary(transactions):
    
    customer_summaries = {}

    for d in transactions:
        if d["customer"] not in customer_summaries:
            customer_summaries[d["customer"]] = {}
            customer_summaries[d["customer"]]["count"] = 1
            customer_summaries[d["customer"]]["total"] = d["amount"]
        else:
            customer_summaries[d["customer"]]["count"] = customer_summaries[d["customer"]].get("count", 0) + 1
            customer_summaries[d["customer"]]["total"] = customer_summaries[d["customer"]].get("total", 0) + d["amount"]
    
    for key, value in customer_summaries.items():
        customer_summaries[key]["average"] = value["total"] / value["count"]

    return customer_summaries


# Optional: use these to test your function
print(customer_summary([
    {"customer": "Ana", "amount": 25},
    {"customer": "Ben", "amount": 50},
    {"customer": "Ana", "amount": 15}
]))

print(customer_summary([
    {"customer": "Mia", "amount": 100},
    {"customer": "Leo", "amount": 50},
    {"customer": "Mia", "amount": 25}
]))

print(customer_summary([]))
```

```python
"""
Python Coding Practice — Challenge 27

Task:
Write a function called `top_customer_summary(transactions)` that returns the
customer with the highest total spending, along with their summary stats.

Requirements:
1. `transactions` is a list of dictionaries
2. Each dictionary has keys: "customer" and "amount"
3. Return a dictionary with exactly these keys:
   - "customer": the customer name
   - "count": number of transactions for that customer
   - "total": total amount spent by that customer
   - "average": average transaction amount for that customer
4. If two customers tie for highest total, return the one who appeared first in the input list
5. Do not round the average
6. If the input list is empty, return None

Examples:
top_customer_summary([
    {"customer": "Ana", "amount": 25},
    {"customer": "Ben", "amount": 50},
    {"customer": "Ana", "amount": 15}
]) -> {
    "customer": "Ana",
    "count": 2,
    "total": 40,
    "average": 20.0
}

Note: Ben has total 50 in that example, so actually Ben should be returned:
{
    "customer": "Ben",
    "count": 1,
    "total": 50,
    "average": 50.0
}

top_customer_summary([
    {"customer": "Mia", "amount": 100},
    {"customer": "Leo", "amount": 50},
    {"customer": "Mia", "amount": 25},
    {"customer": "Leo", "amount": 75}
]) -> {
    "customer": "Mia",
    "count": 2,
    "total": 125,
    "average": 62.5
}

Because Mia and Leo tie at 125, and Mia appears first.

top_customer_summary([]) -> None

Write your code below.
"""


def top_customer_summary(transactions):

    if transactions == []:
        return None
    
    customer_summaries = {}

    for d in transactions:
        if d["customer"] not in customer_summaries:
            customer_summaries[d["customer"]] = {}
            customer_summaries[d["customer"]]["count"] = 1
            customer_summaries[d["customer"]]["total"] = d["amount"]
        else:
            customer_summaries[d["customer"]]["count"] = customer_summaries[d["customer"]].get("count", 0) + 1
            customer_summaries[d["customer"]]["total"] = customer_summaries[d["customer"]].get("total", 0) + d["amount"]

    customer_summaries = dict(sorted(customer_summaries.items(), key=lambda item: item[1]["total"], reverse=True))
    customer, summary = next(iter(customer_summaries.items()))
    
    top_customer_summary = {}
    top_customer_summary["customer"] = customer
    top_customer_summary["count"] = summary["count"]
    top_customer_summary["total"] = summary["total"]
    top_customer_summary["average"] = top_customer_summary["total"] / top_customer_summary["count"]

    return top_customer_summary


# Optional: use these to test your function
print(top_customer_summary([
    {"customer": "Ana", "amount": 25},
    {"customer": "Ben", "amount": 50},
    {"customer": "Ana", "amount": 15}
]))

print(top_customer_summary([
    {"customer": "Mia", "amount": 100},
    {"customer": "Leo", "amount": 50},
    {"customer": "Mia", "amount": 25},
    {"customer": "Leo", "amount": 75}
]))

print(top_customer_summary([]))
```
